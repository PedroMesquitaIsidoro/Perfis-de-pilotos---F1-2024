```python
import fastf1
import os

output_folder = 'laps_2024_all'
os.makedirs(output_folder, exist_ok=True)

schedule = fastf1.get_event_schedule(2024)

for _, race in schedule.iterrows():
    gp_name = race['EventName'].split(' Grand Prix')[0]
    print(f'Baixando dados da corrida: {gp_name} 2024')

    try:
        session = fastf1.get_session(2024, gp_name, 'R')
        session.load()

        laps = session.laps  # pega todos os laps, sem filtro

        filename = f"{output_folder}/{gp_name.lower().replace(' ', '_')}_2024_race_laps.csv"
        laps.to_csv(filename, index=False)
        print(f"Salvo em: {filename}")

    except Exception as e:
        print(f"Erro no GP {gp_name}: {e}")
```
