```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# 1. Carregar dados
df = pd.read_csv("df_all_corridas_2024.csv")

# 2. Features que descrevem o estilo de pilotagem
features = [
    "MediaTempo", "DesvioPadrao",
    "NumParadasBox", "DuracaoMediaStint", "PropVoltasMacios",	"PropVoltasMedios",	"PropVoltasDuros"
]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Escolher apenas as duas features
features_2D = ["DesvioPadrao", "PropVoltasMacios"]

# Escalar dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pilotos[features_2D])

# Avaliar K com Silhouette e Cotovelo
sil_scores = []
inertias = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    sil_scores.append(silhouette_score(X_scaled, labels))
    inertias.append(kmeans.inertia_)

# --- Plot do método do cotovelo ---
plt.figure(figsize=(6,4))
plt.plot(K_range, inertias, 'o-')
plt.xlabel("Número de clusters (k)")
plt.ylabel("WCSS")
#plt.title("Método do Cotovelo para escolha de k")
#plt.grid(True)
plt.show()

# --- Plot do método da silhueta ---
plt.figure(figsize=(6,4))
plt.plot(K_range, sil_scores, 'o-')
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette Score")
#plt.title("Escolha do melhor k pelo método da silhueta")
#plt.grid(True)
plt.show()

# Escolher o melhor k
best_k = K_range[np.argmax(sil_scores)]
print(f"Melhor k segundo silhueta: {best_k}")

# Rodar KMeans com o melhor k
kmeans = KMeans(n_clusters=best_k, random_state=0)
df_pilotos["Cluster"] = kmeans.fit_predict(X_scaled)

# Visualizar clusters em 2D
plt.figure(figsize=(8,6))
for cluster in range(best_k):
    idx = df_pilotos["Cluster"] == cluster
    plt.scatter(X_scaled[idx,0], X_scaled[idx,1], label=f'Cluster {cluster}', s=80)

# Adicionar nomes dos pilotos
for i, piloto in enumerate(df_pilotos["Driver"]):
    plt.text(X_scaled[i,0]+0.05, X_scaled[i,1]+0.05, piloto, fontsize=8)

plt.xlabel("Desvio Padrão do Tempo de Volta")
plt.ylabel("Proporção de Voltas com Pneus Macios")
#plt.title("Clusterização dos pilotos (2 features)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionar as features e normalizar (se ainda não estiverem)
X_scaled = df_pilotos[features].values

# Calcular linkage (método de Ward)
Z = linkage(X_scaled, method='ward')

# --- Dendrograma ---
plt.figure(figsize=(10, 6))
dendrogram(Z, labels=df_pilotos['Driver'].values, leaf_rotation=90, leaf_font_size=10)
#plt.title("Dendrograma dos Pilotos (Ward)")
plt.xlabel("Pilotos")
plt.ylabel("Distância Euclidiana")
plt.show()

# --- Matriz de distâncias ---
# Calcula as distâncias euclidianas entre todos os pilotos
dist_matrix = pdist(X_scaled, metric='euclidean')

# Converte para forma quadrada (n x n)
dist_square = squareform(dist_matrix)

# Cria o heatmap da matriz de distâncias
plt.figure(figsize=(8, 6))
sns.heatmap(dist_square, xticklabels=df_pilotos['Driver'].values,
            yticklabels=df_pilotos['Driver'].values, cmap='viridis')
#plt.title("Matriz de Distâncias Euclidianas entre Pilotos")
plt.xlabel("Pilotos")
plt.ylabel("Pilotos")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 5. Escolher o melhor k usando o método do cotovelo
X_scaled = df_pilotos[features].values
inertia = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot do método do cotovelo
plt.figure(figsize=(6,4))
plt.plot(K, inertia, 'o-')
plt.xlabel("Número de clusters (k)")
plt.ylabel("WCSS")
#plt.title("Escolha do melhor k pelo cotovelo")
plt.show()

# 6. Rodar K-Means com o k escolhido manualmente pelo cotovelo
best_k = 4  # exemplo, escolha visual a partir do gráfico do cotovelo
kmeans = KMeans(n_clusters=best_k, random_state=52)
df_pilotos["Cluster"] = kmeans.fit_predict(X_scaled)

# 7. Ver médias por cluster
print(df_pilotos.groupby("Cluster")[features].mean())

# 8. Visualizar clusters em 2D com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for cluster in range(best_k):
    idx = df_pilotos["Cluster"] == cluster
    plt.scatter(X_pca[idx, 0], X_pca[idx, 1], label=f'Cluster {cluster}', cmap="tab10")

for i, piloto in enumerate(df_pilotos["Driver"]):
    plt.text(X_pca[i, 0], X_pca[i, 1], piloto, fontsize=8)

plt.xlabel("PC1")
plt.ylabel("PC2")
#plt.title("Clusterização dos pilotos (PCA)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def biplot(score, coeff, labels=None, cluster_labels=None):
    xs = score[:,0]
    ys = score[:,1]
    plt.figure(figsize=(8,6))

    # plotar cada cluster separadamente para criar legenda
    unique_clusters = np.unique(df_pilotos["Cluster"])
    for cluster in unique_clusters:
        idx = df_pilotos["Cluster"] == cluster
        plt.scatter(xs[idx], ys[idx], s=60, label=f'Cluster {cluster}', cmap="tab10")

    # adicionar nomes dos pilotos
    for i, piloto in enumerate(df_pilotos["Driver"]):
        plt.text(xs[i]+0.02, ys[i]+0.02, piloto, fontsize=8)

    # desenhar vetores das variáveis (setas + texto organizado)
    for i in range(coeff.shape[0]):
        x = coeff[i,0]
        y = coeff[i,1]

        # seta
        plt.arrow(0, 0, x, y, color="r", alpha=0.5, head_width=0.05)

        # deslocamento automático
        offset_x = 0.04 if x >= 0 else -0.04
        offset_y = 0.04 if y >= 0 else -0.04

        nome = labels[i] if labels is not None else f"Var{i+1}"

        plt.text(
            x + offset_x,
            y + offset_y,
            nome,
            color="g",
            fontsize=7,
            ha="left" if x >= 0 else "right",
            va="bottom" if y >= 0 else "top"
        )

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid()
    plt.legend()
    plt.show()


# Rodar PCA de novo para pegar loadings
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

biplot(X_pca, pca.components_.T, labels=features)
```

```python
# 5. Escolher o melhor k usando Silhouette
X_scaled = df_pilotos[features].values
scores = []
K = range(2, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append(score)

# Plot Silhouette Score
plt.figure(figsize=(6,4))
plt.plot(K, scores, 'o-')
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette Score")
#plt.title("Escolha do melhor k")
plt.show()

# 6. Rodar K-Means com o k escolhido
best_k = K[scores.index(max(scores))]
kmeans = KMeans(n_clusters=best_k, random_state=0)
df_pilotos["Cluster"] = kmeans.fit_predict(X_scaled)

# 7. Ver médias por cluster
print(df_pilotos.groupby("Cluster")[features].mean())

# 8. Visualizar clusters em 2D com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))

for cluster in range(best_k):
    idx = df_pilotos["Cluster"] == cluster
    plt.scatter(X_pca[idx, 0], X_pca[idx, 1], label=f'Cluster {cluster}', cmap="tab10")

for i, piloto in enumerate(df_pilotos["Driver"]):
    plt.text(X_pca[i, 0], X_pca[i, 1], piloto, fontsize=8)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusterização dos pilotos (PCA)")
plt.legend()  # <-- legenda aparece aqui
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Pegando os centróides do K-Means (já em escala)
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=features)

# Normalizar os valores entre 0 e 1 para radar (deixa o visual mais limpo)
centroids_norm = (centroids - centroids.min()) / (centroids.max() - centroids.min())

# Número de variáveis
num_vars = len(features)

# Ângulos para o gráfico
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # fecha o círculo

# Cores fixas para cada cluster (mesmas do texto)
cluster_colors = {
    0: '#1f77b4',  # Azul
    1: '#ff7f0e',  # Laranja
    2: '#2ca02c',  # Verde
    3: '#d62728',  # Vermelho
}

# Criar gráfico para cada cluster
for i, row in centroids_norm.iterrows():
    values = row.tolist()
    values += values[:1]  # fecha o círculo

    fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
    color = cluster_colors.get(i, 'gray')  # usa a cor padrão se não achar

    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid', label=f'Cluster {i}')
    ax.fill(angles, values, color=color, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(features, fontsize=9)
    ax.set_yticklabels([])
    #ax.set_title(f'Radar dos atributos - Cluster {i}', size=12, y=1.1)

    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    plt.tight_layout()
    plt.show()
```
